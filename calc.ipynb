# calcolo costi: asporto vs cucina

## calcolo costi cucina

### costi una tantum

```python
COSTI_UNA_TANTUM = {
    "COSTRUZIONE" : 10_000,
    "VOLTURA"     : 125           # costo per voltura (1,5 kW in piu), permette di usare altri eletrodomestici mentre cucini estimato 120 - 130 euro
}
```

### consumi periodici

#### eletrici in kWh

```python
ORE_DI_UTILIZZO_BIMBY = 70 # ore anuali

COSTI_ELETRICI = {
    "PIASTRA"       : 475,          # 450 - 500
    "FRIGRORIFERO"  : 600,          
    "FORNO"         : 120,          
    "LAVATRICE"     : 260,
    "BIMBY"         : 0.5 * ORE_DI_UTILIZZO_BIMBY,
    "ALTRO"         : 0.5 * 365
}

COSTI_ELETRICI
```

#### altri costi in eruo

```python
COSTO_KILOWATT_ORA = 0.25
COSTI_RICORRENTI = {
    "MANTENIMENTO_VOLTURA" : 35,           # per continuare ad avere la voltura
    "SPESA" : 5784,                        # spesa media in italia in piemonte 2022
    "EQUIPAGGIAMENTO" : 60,                # Nuove pentole, accessori, ecc.
    "COSTO_MEDIO_RIPARAZIONI_ANUALE" : 80, # costo medio delle riparazioni annuale
    "COSTI_ELETRICI" : sum(COSTI_ELETRICI.values()) * COSTO_KILOWATT_ORA
}
```

### costi totali

```python
f"Costi una tantum: {sum(COSTI_UNA_TANTUM.values())}€ + Costi anuali: {sum(COSTI_RICORRENTI.values())}€"
```

# costi asporto

```python
# Prezzi leggermente inferiori dato che mangi vegetariano
LUCIA = {
    "COSTO_COLAZIONE" : 4,
    "COSTO_ASPORTO_PRANZO" : 7,
    "COSTO_ASPORTO_CENA" : 9
}

FRANCESCO = {
    "COSTO_COLAZIONE" : 0,
    "COSTO_ASPORTO_PRANZO" : 8,
    "COSTO_ASPORTO_CENA" : 10
}

MASSIMILIANO = {
    "COSTO_COLAZIONE" : 1.5,
    "COSTO_ASPORTO_PRANZO" : 3.7,
    "COSTO_ASPORTO_CENA" : 3.7
}

COSTO_ASPORTO_ANUALE = (sum(LUCIA.values()) + (sum(FRANCESCO.values()) + sum(MASSIMILIANO.values())) / 2) * 365
f"Costo Anuale asporto {COSTO_ASPORTO_ANUALE}€"
```

Costo aggiuntivo per vivere d'asporto

```python
f"Costi aggiuntivi: {abs(sum(COSTI_RICORRENTI.values()) - COSTO_ASPORTO_ANUALE)}€"
```

Risparmio Tempistco

```python
TEMPO_COLAZIONE = 5
TEMPO_PRANZO = 30
TEMPO_CENA = 30

TEMPO_RISPARMIO = ((TEMPO_COLAZIONE + TEMPO_PRANZO + TEMPO_CENA) * 365) / 60
f"Risparmio tempo anuale: {TEMPO_RISPARMIO} ore o {TEMPO_RISPARMIO / 24} giorni"
```
